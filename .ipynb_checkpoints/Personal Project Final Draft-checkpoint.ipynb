### Background and Data Decisions

The sport of Formula 1 is primarily located within Europe despite having a global championship that involves many international fans. Despite international outreach, the sport still has a limited amount of drivers that are from other continents. Due to this, when there are minorities, they are often backed largely by people from the same race or ethnicity. Additionally, they may be treated differently in a negative manner due to racism and xenophobia. Due to this, I wanted to see how minorities are treated within the sport of Formula 1 through their headline word choices. 

For this project, I analyzed the headlines of the 5 minority drivers within the 20 on the formula 1 grid currently. They range in performance throughout the sport and are of different ethnicities each. All headlines utilized for this project hold comments about them either made on their performance, statements on their own performance, or statements about their performance by others within their team's personnel. Any headlines that are about other people or composite performance of the team are removed for the sake of accuracy.

### Leading Question

How are minority drivers represented in Formula 1 headlines? Are the words used more positive or negative?

### Function Creation
****

First, I made a function to clean the data for each text file necessary. The function opens the file, removes punctuation and makes each letter lowercase. From there, tokens are made from the txt files. Regular stop words are then removed from the data, but additional stopwords are removed as necessary. These stop words are sorted by category for my own editing purposes, first by driver, then by general sport terminology, then other words that were commonly showing up between all headlines. Once stop words are removed from the tokens, the tokens are joined together in order to make lemmas and return them for further analysis. 

```python
def data_cleaner(file_name):
    file = open(file_name, encoding = 'utf-8').read()
    new_punctuation = punct + '‘' + '’' + '–'
    for char in new_punctuation:
        file = file.lower().replace(char, "")
    tokens = file.split()
    stop = stopwords.words('english')
    hamilton_stops = ['lewis', 'hamilton', 'hamiltons', 'nico', 'rosberg', '40th', 'birthday', 'toto', 'wolff', 'jenson', 'button']
    perez_stops = ['christian', 'horner', 'red', 'bull', 'sergio', 'perez', 'rb20']
    albon_stops = ['alex', 'albon', 'williams', 'logan', 'sargeant', 'sargeants', 'magnussen', 'helmut', 'marko', 'red', 'bull', 'kevin']
    tsunoda_stops = ['yuki', 'tsunoda', 'rb', 'laurent', 'mekies', 'rbs', 'engineers', 'daniel', 'ricciardo']
    zhou_stops = ['zhou', 'guanyu', 'bottas', 'sauber', 'kick']
    formula1_stops = ['2024', '2025', 'season', 'car', 'team', 'mate', 'track', 'canada','monaco', 'miami', 'china', 'japan', 'japanese', 'australia', 'imola', 'australian', 'melbourne', 'bahrain', 'gp', 'q1', 'q2', 'q3', 'fp2', 'f1', 'postf1', 'mercedes', 'ferrari', 'p2', 'lap', 'damon', 'hill', 'max', 'verstappen', 'saudi', 'arabia', 'shanghai', 'sprint', 'race', 'qualifying', 'friday', 'saturday', 'sunday', 'suzuka', 'turn', '1', 'ferraris', 'bulls', 'spanish', 'wheel', 'nut', 'spare,', 'chassis', 'fp1', 'pit', 'stop']
    extra_stops = ['whats', 'ive', 'weve', 'got', 'let', 'hes']
    removed_stopwords = [word for word in tokens if word not in stopwords.words('english') + hamilton_stops + perez_stops + albon_stops + tsunoda_stops + zhou_stops + formula1_stops + extra_stops]
    words = ' '.join(removed_stopwords)
    doc = nlp(words)
    return [token.lemma_ for token in doc]
```

After cleaning the data, an additional function was made to sort the lemmas by partrs of speech. From there, adjectives and verbs were picked out from the data. To review the parts of speech, the tags are returned instead of the labels created. This allows further cleaning that is precise to the specific txt file being used, as later defined in the document. 

```python
def tag_maker(cleaned_data, list_name):
    pos_tags = pos_tag(cleaned_data)
    for item in pos_tags:
        if (item[1] == ('JJ'or 'VB' or 'VBD')):
            list_name.append(item[0])
    return print(pos_tags)
```

Although an additional step is needed between the last function and this one, showcased below, due to specific word sets utilized, this is sorted per each section before moving on to the next step. 

After manual sorting, the processed data for each txt file can be graphed adequately. In order to showcase how often the specific words show up to show any correlation, a dataframe is created, then subsequently graphed. Both the dataframe and graph are showcased from the function, such that both pieces of information can be utilized. 

However, I discovered after the first analysis that this frequency graph is not particularly useful for anything other than preserving original material within the analysis. As such, I decided to look into other methods of analysis. 

```python
def frequency_graph(processed_words):
    frequency = Counter(processed_words)
    frequency_graph = pd.DataFrame(frequency.most_common(30), columns = ['Word', 'Frequency'])
    print(frequency_graph)
    return (sns.barplot(frequency_graph, x = 'Frequency', y = 'Word', orient = 'h'))
```

Due to the fact that the frequency graph was not helpful, I decided to assign values of affect within the lists of words available.

Similar to the last step, additional data processing is done per data set manually. This is also done before processing the next step for each data set. This is due to the needed manual input for how each word is utilized within context of the sport. 

After processing the data, graphs that showcase the general affect within the word used. These tones highlight similar word choice, even if the words are not necessarily the same. In allowing the affect to be visualized in a multitude of ways, this allows for comparison between each driver and a full understanding of the tone of the headlines amongst them.

```python
def affect_graphs(affect_values):
    df = pd.DataFrame(affect_values)
    affect_count = df['Affect'].value_counts().sort_index()
    affect_plot = df['Affect'].value_counts().sort_index().plot.area()
    affect_bar = df['Affect'].value_counts().sort_index().plot(kind = 'bar', color = ['#eb9191', '#b2b7f1', '#90cc98'])
    return print(affect_count, affect_plot, affect_bar)
```
